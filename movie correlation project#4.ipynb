```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# missing data
for col in df.columns:
    p_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,p_missing))
    
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#
df['year_corrected'] = df['released'].astype(str).str[:4]
df.head(10)
```

```python
df.sort_values(by=['gross'],inplace = False , ascending= True)
df.head(10)
```

```python
# drop duplicates
df.drop_duplicates()
```

```python
# correlating
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budet vs Gross')
plt.xlabel('buget')
plt.ylabel('gross')
plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"orange"},line_kws={"color":"blue"})
```

```python
df.corr(method='pearson')# other are kendall,spearman,pearson
```

```python
#high correlation btw budget and gross
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix,annot = True)
plt.title('Correlation matrix for matrix Features')
plt.xlabel('movie feature')
plt.ylabel('movie features')
plt.show()
```

```python
df
```

```python
df_numerized = df
```

```python
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
```

```python
df_numerized
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix,annot = True)
plt.title('Correlation matrix for matrix Features')
plt.xlabel('movie feature')
plt.ylabel('movie features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
